```python
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x)) #exp
```

```python
x = np.array([-1.0 , 2.0, 2.0])
print(sigmoid(x))
```

```python
t = np.array([1.0, 2.0, 3.0])
print(1.0 + t)
print(1.0 / t)
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
```

```python
plt.plot(x,y)
plt.ylim(-0.1, 1.1) # y축 범위 지정
plt.show
```
