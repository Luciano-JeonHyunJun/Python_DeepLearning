```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def step_function(x):
    return np.array(x > 0, dtype=np.int)
```

```python
x = np.arange(-5.0, 5.0, 0.1) #np.arange는 -5.0에서 5.0전까지 0.1 단위로 배열 생성
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) #y축의 범위 지정
plt.show
```
